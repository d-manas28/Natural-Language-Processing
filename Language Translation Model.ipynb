```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/drive/MyDrive/__MyProjects @IITP/Files/ver1_hindiwords_list.csv')
df
```

```python
list_of_words = df['H_words'].tolist()
list_of_words
```

```python
print(type(list_of_words))
```

```python
from googletrans import Translator
translatorModel = Translator()
```

```python
res = translatorModel.translate(list_of_words[0],src='hi',dest='en')
print(res.text)
```

```python
translated_words = []
for i in list_of_words:
  res = translatorModel.translate(i,src='hi',dest='en')
  translated_words.append(res.text)
```

```python
translated_words
```

```python
print(len(translated_words))
```

```python
res_df = pd.DataFrame()
res_df['Hindi Words'] = list_of_words
res_df['Google Translation'] = translated_words
```

```python
print(res_df)
```
