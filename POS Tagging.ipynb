>Let's calculate through viterbi algorithm

```python
def viterbi_algo(sentence):
  sentence = sentence.replace(",","")
  sentence = sentence.replace(".","")
  sentence = sentence.replace("`","")
  sentence = sentence.replace("'","")
  sentence = sentence.replace("?","")
  words = sentence.split(" ")

  df_words = pd.DataFrame()
  tag_list = list(tag_set)
  df_words['Tags'] = tag_list
  res_tag_list = []
  for  i in words:
    df_words[i] = [1.0 for i in range(len(tag_set))]

  df_words.set_index("Tags", inplace = True)

  flag = False
  start_word = words[0]
  if(start_word not in name_set):
    flag = True
  for i in range(len(tag_set)):
    if flag == False:
      em_prob = df.loc[start_word][tag_list[i]]
    else:
      em_prob = 1
    t_prob = df_tp.loc['S'][tag_list[i]]
    res_prob = em_prob * t_prob

    # print("Prob for ",start_word,"and Tag: ",tag_list[i],"EmP:",em_prob," TrP:",t_prob," Res: ",res_prob)
    df_words.loc[tag_list[i]][start_word] = df_words.loc[tag_list[i]][start_word] * res_prob
  
  for i in range(1,len(words)):
    flag = False
    # find maximum prob of prev col and tag associated with it. Add the tag into result list
    curr_word = words[i]
    if(curr_word not in name_set):
      flag = True
    prev_col = df_words.columns[i-1]
    last_res_prob = df_words[prev_col].max()
    last_res_tag = df_words[prev_col].idxmax()
    res_tag_list.append(last_res_tag)

    for j in range(len(tag_list)):
      if(flag == False):
        curr_em_prob = df.loc[curr_word][tag_list[j]] 
      else:
        curr_em_prob = 1    
      curr_t_prob = df_tp.loc[last_res_tag][tag_list[j]]
      res_prob = curr_em_prob * curr_t_prob * last_res_prob
      df_words.loc[tag_list[j]][curr_word] = df_words.loc[tag_list[j]][curr_word] * res_prob  

  # print(df_words)
  return [res_prob,words,res_tag_list]
```

```python
res_prob,words,res_tag_list = viterbi_algo("Sales rose 10% to $13.8 billion from $12.5 billion.")
```
